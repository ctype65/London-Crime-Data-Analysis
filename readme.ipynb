# London Crime Data Analysis

Area-wise crime analysis of London to see nau pattern in different types of crime and get to know the neighborhood which can help to understand where the most and least crime occur in London which can be helpful for residence and places with most crime rates can help local police to control criminal activities.



The dataset can be downloaded from Kaggle which has a strong data science community with powerful tools and resources, [click here to download](https://www.kaggle.com/jboysen/london-crime)
